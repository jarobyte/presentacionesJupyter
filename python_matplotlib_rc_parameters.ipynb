# Q1. K-Nearest Neighbors Classifier on Synthetic Data

## (a) Create the dataset

```python
x = np.arange(-1.5, 1.6, 0.1)
x[:10]
```

```python
y = np.arange(-1.5, 1.6, 0.1)
y[:10]
```

```python
grid = np.array([[(x_value, y_value) for y_value in y] for x_value in x]).reshape(-1, 2)
grid[:5]
```

```python
grid.shape
```

```python
def p_norm(a, p):
    return (abs(a) ** p).sum(axis = 1) ** (1/p)
```

```python
p_norm(grid, 2).shape
```

To see that the implementation of the "p" norms is correct, we can plot the unitary disc for some values of p and see if the plot is correct.

Since we are setting the "c" parameter (color parameter) to plot an array of True's and False's, the unitary disk is actually **not** plotted.

```python
p = 1
plt.scatter(x = grid[:, 0], y = grid[:, 1], c = p_norm(grid, p) <= 1)
plt.title("Unitary disk in {}-norm".format(p))
plt.show()
```

```python
p = 2
plt.scatter(x = grid[:, 0], y = grid[:, 1], c = p_norm(grid, p) <= 1)
plt.title("Unitary disk in {}-norm".format(p))
plt.show()
```

```python
p = 100
plt.scatter(x = grid[:, 0], y = grid[:, 1], c = p_norm(grid, p) <= 1)
plt.title("Unitary disk in {}-norm".format(p))
plt.show()
```

We can see that the implementation of the p norms seems correct.
