```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize'] = (10, 6)

# This parameter gets in conflict with sns.set()
matplotlib.rcParams['font.size'] = 20
```

```python
x = np.arange(-1.5, 1.6, 0.1)
x[:10]

y = np.arange(-1.5, 1.6, 0.1)
y[:10]

grid = np.array([[(x_value, y_value) for y_value in y] for x_value in x]).reshape(-1, 2)
grid[:5]

def p_norm(a, p):
    return (abs(a) ** p).sum(axis = 1) ** (1/p)
```

```python
p = 1
plt.scatter(x = grid[:, 0], y = grid[:, 1], c = p_norm(grid, p) <= 1)
plt.title("Unitary disk in {}-norm".format(p))
plt.show()
```

```python
p = 2
plt.scatter(x = grid[:, 0], y = grid[:, 1], c = p_norm(grid, p) <= 1)
plt.title("Unitary disk in {}-norm".format(p))
plt.show()
```
