```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.patches as mpatches
```

```python
x = np.arange(-1.5, 1.6, 0.1)
x[:10]

y = np.arange(-1.5, 1.6, 0.1)
y[:10]

X = np.array([[(x_value, y_value) for y_value in y] for x_value in x]).reshape(-1, 2)
X[:5]

X.shape

classes = (np.linalg.norm(X, ord = 2, axis = 1) >= 1) + 1
classes[:5]
```

```python
plt.scatter(x = X[:, 0], 
            y = X[:, 1], 
            c = ["orange" if c == 1 else "blue" for c in classes])
plt.legend(handles = [mpatches.Patch(color = 'orange', label = "Class 1"),
                      mpatches.Patch(color = 'blue', label = "Class 2")])
plt.title("The training set")
plt.xlabel("X coordinate")
plt.ylabel("Y coordinate")

plt.show()
```
